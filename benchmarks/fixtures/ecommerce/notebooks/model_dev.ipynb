# Order value model — dev notebook
Predicting order amount from account tenure. Cross-validated, should be solid before handoff.

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_score

orders = pd.read_csv('../data/orders.csv', parse_dates=['order_date'])
users = pd.read_csv('../data/users.csv', parse_dates=['signup_date'])
df = orders.merge(users, on='user_id')
df['tenure_days'] = (df['order_date'] - df['signup_date']).dt.days
```

```python
X = df[['tenure_days']]
y = df['amount']

kf = KFold(n_splits=5, shuffle=True)
model = GradientBoostingRegressor()
scores = cross_val_score(model, X, y, cv=kf, scoring='neg_root_mean_squared_error')
print('CV RMSE:', -scores.mean())
```

Looks stable across folds — ready to hand off?
